## 3. Análisis regresión

```python
#

GasLP_df3['cens'] = 0
GasLP_df3.loc[(GasLP_df3.lgasto_gaslp == 0), 'cens'] = -1
GasLP_df3
```

```python
#
# Variable de censura
cens = pd.Series(GasLP_df3.cens)

cens.value_counts(normalize = True)
```

```python
# Definimos variables
y = GasLP_df3.lgasto_gaslp
x = GasLP_df3[['ling_cor', 'ling_cor2', 'tot_integ', 'lp_gaslp', 'acc_gasnat']]
```

```python
# Estimación del modelo Tobit

tr = TobitModel()
tr = tr.fit(x, y, cens, verbose = True)
```

```python
# Resultados
tr.intercept_, tr.coef_
```

```python
tr.sigma_
```

```python
# Adecuaciones a X:
x = sm.add_constant(x)
```

```python
# OLS
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
```

```python
# Resultadoss de Tobit:
tr.coef_
```

```python
# Comparemos:
# Tobit:

tr.intercept_, tr.coef_[0], tr.coef_[1], tr.coef_[2], tr.coef_[3], tr.coef_[4]
```
